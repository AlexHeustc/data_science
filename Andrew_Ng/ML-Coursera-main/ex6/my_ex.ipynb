```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from scipy.io import loadmat
```

```python
raw_data = loadmat("data/ex6data1.mat")
data = pd.DataFrame(raw_data.get('X'),columns=['X1','X2'])
data['y']=raw_data.get('y')
data.head()
```

```python
def plot_init_data(data,fig,ax):
    positive = data[data['y'].isin([1])]
    negative = data[data['y'].isin([0])]

    ax.scatter(positive['X1'],positive['X2'],c='r',marker='o',label='Positive')
    ax.scatter(negative['X1'],negative['X2'],c='b',marker='x',label='Negative')
```

```python
fig,ax = plt.subplots(figsize=(20,8))
plot_init_data(data,fig,ax)
plt.legend()
plt.show()
```
