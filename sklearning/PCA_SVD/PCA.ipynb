```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
y = iris.target
x = iris.data
```

```python
x.shape
import pandas as pd
pd.DataFrame(x).head()
```

```python
y
```

```python
pca = PCA(n_components=2)
x_dr = pca.fit_transform(x)
x_dr
```

```python
x_dr[y == 0,0]

color = ['red', 'black', 'orange']
iris.target_names

plt.figure()
for i in [0,1,2]:
    plt.scatter(
        x_dr[y == i, 0],
        x_dr[y == i, 1],
        alpha=0.7,
        c=color[i],
        label = iris.target_names[i]
    )
plt.legend()
plt.title('PCA of iris')
plt.show()
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
import numpy as np
pca_line = PCA().fit(x)
plt.plot([1,2,3,4],np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks([1,2,3,4])
plt.xlabel("number of components after dimension reduction")
plt.ylabel("cumulative explained variance ratio")
plt.show()
```

```python
pca_mle = PCA(n_components="mle")
pca_mle = pca_mle.fit(x)
x_mle = pca_mle.transform(x)
x_mle
#可以发现，mle为我们自动选择了3个特征
pca_mle.explained_variance_ratio_.sum()
```

```python
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
face = fetch_lfw_people(min_faces_per_person=60)
face.images.shape
face.data.shape
x = face.data
face
```

```python
fig,axes = plt.subplots(4,5,
                       figsize=(8,4),
                       subplot_kw = {"xticks":[],"yticks":[]}
                       )
fig
axes
axes.shape
```

```python
axes.flat
enumerate(axes.flat)
```

```python
fig,axes = plt.subplots(4,5,
                       figsize=(8,4),
                       subplot_kw = {"xticks":[],"yticks":[]}
                       )

for i,ax in enumerate(axes.flat):
    ax.imshow(face.images[i,:,:]
              ,cmap = 'gray'
              )

```

```python
pca = PCA(150).fit(x)
V = pca.components_
V.shape
```

```python
fig, axes = plt.subplots(3,8,figsize=(8,4),subplot_kw = {"xticks":[],"yticks":[]})
for i, ax in enumerate(axes.flat):
    ax.imshow(V[i,:].reshape(62,47),cmap="gray")
```

```python
x_dr = pca.fit_transform(x)
x_inverse = pca.inverse_transform(x_dr)
x_inverse.shape

fig,ax = plt.subplots(2,10,figsize = (10,2.5),subplot_kw={"xticks":[],"yticks":[]})
for i in range(10):
    ax[0,i].imshow(face.images[i,:,:],cmap="binary_r")
    ax[1,i].imshow(x_inverse[i].reshape(62,47),cmap="binary_r")
```

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
digits = load_digits()
digits.data.shape
```

```python
def plot_digits(data):
    fig,axes = plt.subplots(4,10
                            ,figsize = (4,10)
                            ,subplot_kw = {"xticks":[],"yticks":[]}
                            )
    for i,ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8,8),cmap = 'binary')

plot_digits(digits.data)
```

```python
np.random.RandomState(42)
noisy = np.random.normal(digits.data,2)
plot_digits(noisy)
```

```python
pca = PCA(0.5).fit(noisy)
X_dr = pca.transform(noisy)
X_dr.shape
```

```python
without_noise = pca.inverse_transform(X_dr)
plot_digits(without_noise)
```
